**En ce qui concerne le web scraping on a travaillé avec le meme principe pour scraper tout les sites .
Donc voila l'architecture Générale du programme qu'on a appliqué en changeant juste les balises qui contient les données pour les autres sites** 


**BeautifulSoup** est la bibliothèque qui extrait des données de fichiers HTML


```
import requests
from bs4 import BeautifulSoup
```

L'envoie d'une requete HTTP 'GET' à l'URL du site web

```
URL = 'http://soaalwajawab.blogspot.com/2014/08/6001-7000.html'
page = requests.get(URL)
```

Récupération du code HTML du site avec Beautiful Soup

```
soup = BeautifulSoup(page.content, 'html.parser')
```

Récupération des données en déterminant les balises et les indices

```
results = soup.find('div',dir='rtl')
```

Récuperer la totalité des balises qui contient les données

```
all = results.findAll('span')
```

Extraire le texte et le diviser en question et réponse

```
all=[q.text.strip() for q in all]
QA = [q.split('؟') for q in all]
```

Sauvegarde des résultats

```
with open('Questions.txt','w',encoding='utf-8') as f:
    for q in QA:
        f.write(q[0].strip())
        f.write('\n')

with open('Réponses.txt','w',encoding='utf-8') as f:
    for q in QA:
        f.write(q[1].strip())
        f.write('\n')
```
